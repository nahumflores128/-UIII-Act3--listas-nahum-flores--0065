```python
# Base de datos global (Lista de diccionarios)
administradores = []

def agregar_administrador():
    nombre = input("Nombre del administrador: ")
    edad = input("Edad del administrador: ")
    departamento = input("Departamento: ")

    administrador = {
        "nombre": nombre,
        "edad": edad,
        "departamento": departamento
    }

    administradores.append(administrador)
    print(f"\n✅ {nombre} ha sido agregado correctamente.")

def leer_administrador(administrador):
    print(f"\n--- Detalle del Administrador ---")
    print(f"Nombre: {administrador['nombre']}")
    print(f"Edad: {administrador['edad']}")
    print(f"Departamento: {administrador['departamento']}")

def listar_administradores():
    if not administradores:
        print("\n⚠️ La lista está vacía.")
    else:
        print("\n--- Lista de Administradores ---")
        for i, administrador in enumerate(administradores, 1):
            print(f"{i}. {administrador['nombre']} ({administrador['departamento']})")

def buscar_administrador():
    nombre_buscado = input("Introduce el nombre del administrador a buscar: ")

    for administrador in administradores:
        if administrador['nombre'].lower() == nombre_buscado.lower():
            leer_administrador(administrador)
            return administrador

    print("\n❌ Administrador no encontrado.")
    return None

def modificar_administrador():
    administrador = buscar_administrador()

    if administrador:
        print("\n--- Editando datos (deja en blanco para no cambiar) ---")

        nuevo_nombre = input(f"Nuevo nombre [{administrador['nombre']}]: ")
        nueva_edad = input(f"Nueva edad [{administrador['edad']}]: ")
        nuevo_departamento = input(f"Nuevo departamento [{administrador['departamento']}]: ")

        if nuevo_nombre:
            administrador['nombre'] = nuevo_nombre

        if nueva_edad:
            administrador['edad'] = nueva_edad

        if nuevo_departamento:
            administrador['departamento'] = nuevo_departamento

        print("\n✨ Datos actualizados.")

def borrar_administrador():
    nombre_borrar = input("Nombre del administrador a eliminar: ")

    for i, administrador in enumerate(administradores):

        if administrador['nombre'].lower() == nombre_borrar.lower():

            confirmacion = input(f"¿Seguro que quieres borrar a {administrador['nombre']}? (s/n): ")

            if confirmacion.lower() == 's':
                administradores.pop(i)
                print("\n🗑️ Administrador eliminado.")
                return

    print("\n❌ No se encontró el administrador.")

# --- Menú Principal con While ---
def menu():

    continuar = True

    while continuar:

        print("\n==============================")
        print(" GESTIÓN DE ADMINISTRADORES ")
        print("==============================")
        print("1. Agregar administrador")
        print("2. Listar administradores")
        print("3. Buscar administrador")
        print("4. Modificar administrador")
        print("5. Borrar administrador")
        print("6. Salir")

        opcion = input("\nSelecciona una opción: ")

        if opcion == "1":
            agregar_administrador()

        elif opcion == "2":
            listar_administradores()

        elif opcion == "3":
            buscar_administrador()

        elif opcion == "4":
            modificar_administrador()

        elif opcion == "5":
            borrar_administrador()

        elif opcion == "6":
            print("¡Hasta luego!")
            continuar = False

        else:
            print("❌ Opción no válida, intenta de nuevo.")

# Ejecutar el programa
menu()
```
